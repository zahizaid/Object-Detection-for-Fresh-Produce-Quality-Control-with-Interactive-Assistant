```python
results=model.train(data=r"C:\Users\zahiz\Desktop\yolo_capstone5\Sayuran 2.0.v12i.yolov11\data.yaml",epochs=50, imgsz=640)
```

```python
# Deploy custom trained model
model_custom = YOLO("runs/detect/train/weights/best.pt")
result = model_custom.predict(r"C:\Users\zahiz\Desktop\yolo_capstone5\Carrot__40927.jpg")
plt.show(result)
```

```python
img = result[0].plot()
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.xticks([])
plt.yticks([])
plt.grid(False)
plt.axis('off')
plt.show()
```

```python
cap = cv2.VideoCapture(0)  # 0 is the default webcam, change if needed

if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()

while True:
    ret, frame = cap.read()  
    if not ret:
        print("Error: Failed to capture frame.")
        break


    results = model_custom.predict(frame)

    
    annotated_frame = results[0].plot()

    cv2.imshow("YOLO Webcam Detection", annotated_frame)

    # Press 'q' to break the loop and close the webcam
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close the window
cap.release()
cv2.destroyAllWindows()
```
